```python
flat_data=[]  #to store data
target=[]  #to store y values
categories=['Cat','Dog']
datadir='/content/drive/MyDrive/Data Set-Luminar/Data (1)'
```

```python
import numpy as np
for i in categories:
  print('Loading categories',i)
  path=os.path.join(datadir,i)
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))
    img_resized=resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten()) #flatten used to convert to one dimensional array
    target.append(categories.index(i))
  print('Loaded category:',i,'sucessfully')
```

```python

df=pd.DataFrame(flat_data)
df
```

```python
df['target']=target
target
df.head()
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=3)
from sklearn.svm import SVC
clf=SVC()
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*1)
```
